```python
import pandas as pd

url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
df = pd.read_csv(url, header=None)
df.columns = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth', 'Species']
print(df.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# NOW YOU FILL IN:

# 1. Basic exploration
print("Shape:", df.shape)  # rows, columns
print("\nData types:")
print(df.dtypes)
print("\nMissing values:")
print(df.isnull().sum())
print("\nBasic statistics:")
print(df.describe())

# 2. Ask your question and answer it
# Example: Which species has largest average sepal length?
print("\n--- MY QUESTION ---")
print("Which iris species has the largest average sepal length?")
print(df.groupby('Species')['SepalLength'].mean())

# 3. Make a visualization
plt.figure(figsize=(8, 5))
df.groupby('Species')['SepalLength'].mean().plot(kind='bar')
plt.title('Average Sepal Length by Species')
plt.ylabel('Sepal Length')
plt.xlabel('Species')
plt.show()
```
